```python
from sklearn.metrics import mean_squared_error
#we want to find the best m based on the range of MSE.
#we write for loop for finding the best m.

m = [1, 2, 3, 4, 5, 6, 7, 8]
mse=[]
x_flatten=30*(np.random.rand(number_of_samples, 1) - 0.5).flatten()
y_flatten=5 * x_flatten + 15 * x_flatten**2 + 2 * x_flatten**3 + noise_scale*np.random.randn(number_of_samples, 1).flatten()
mymodel = np.poly1d(np.polyfit(x_flatten, y_flatten, 3))

for i in m:
    
    mymodel=np.poly1d(np.polyfit(x_flatten, y_flatten, i))
    y_pred=mymodel(x_flatten)
    mse_value=mean_squared_error(y_flatten,y_pred)
    mse.append(mse_value)
    
    
    
mse
```

```python
plt.plot(m,mse)
plt.title('m-mse')
plt.xlabel('m')
plt.ylabel('MSE')

```

### 3)Change variable noise_scale to 200, 300, 400, 600, 800, 1000 respectively, re-run the algorithm and plot the polynomials with the m found in 2). Discuss the impact of noise scale to the accuracy of the returned parameters.  
[Note: You need to plot a figure like in 1) for EACH choice of noise_scale.] 

```python
number_of_samples = 50
#x = 30*(np.random.rand(number_of_samples, 1) - 0.5)

noise_scale = [200, 300, 400,600, 800, 1000]

#we chose m=3 for re-running the algorithm 
x_flatten=30*(np.random.rand(number_of_samples, 1) - 0.5).flatten()
x_fit=np.linspace(x_flatten.min(), x_flatten.max(), 100)

for i in noise_scale:
    
    y_flatten=5 * x_flatten + 15 * x_flatten**2 + 2 * x_flatten**3 + i*np.random.randn(number_of_samples, 1).flatten()
    mymodel=np.poly1d(np.polyfit(x_flatten, y_flatten,3))
    y_pred=mymodel(x_fit)
    plt.scatter(x_flatten, y_flatten, label=f'Noise Scale: {i}', alpha=0.5)
    
    plt.plot(x_fit, y_pred, linewidth=2) #we plot based on the different noises.
    plt.title('x-y different noise scale')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.legend()
    
    

    
```

### Based on the plots, when we increase the noise, the polynomial curve will deviate from the actual data points. Also, this noise increaseing will lead to variance increasing in the results.  This nosie increasing will make coefients less reliable.  Becuase in this situation the curve will be influnced by these noises rather than the trend in the actual data points.

4) Change variable number_of_samples to 40, 30, 20, 10 respectively, re-ran the algorithm and plot the polynomials with the m found in 2). Discuss the impact of the number of samples to the accuracy of the returned parameters. [Note: You need to plot a figure like in 1) for EACH choice of number_of_samples.]

```python
noise_scale=200


sample = [10, 20, 30,40] #range of sample size

x_flatten=30*(np.random.rand(number_of_samples, 1) - 0.5).flatten()
x_fit=np.linspace(x_flatten.min(), x_flatten.max(), 100)

for i in sample:
    
    x_flatten=30*(np.random.rand(i, 1) - 0.5).flatten()
    x_fit=np.linspace(x_flatten.min(), x_flatten.max(), 100)
    y_flatten=5 * x_flatten + 15 * x_flatten**2 + 2 * x_flatten**3 + noise_scale*np.random.randn(i, 1).flatten()
    mymodel=np.poly1d(np.polyfit(x_flatten, y_flatten,3))
    y_pred=mymodel(x_fit)
    plt.scatter(x_flatten, y_flatten, alpha=0.5, label=f'Sample Size: {i}')
    plt.plot(x_fit, y_pred, linewidth=2) #plot the data for different sample sizes.
    plt.title('x-y different sample size')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.legend()

    

    
```

### Based on the plot, by increasing the sample size, the model coefients will be more accurate. and the model will be not affected by random noise or outliers.Because there are more information and by using this information they can capture the pattern in the data.Also, by increasing the sample, the model will be not overfitted.
